```python
Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.
    DDL (Data Definition Language)

DDL is a category of SQL commands used to define, modify, and manage the structure of database objects such as tables, schemas, and indexes.

DDL changes the database structure, not the actual data manipulation process.

Common DDL Commands:

CREATE
ALTER
DROP
TRUNCATE

Used in databases like MySQL, PostgreSQL, and Oracle Database.

1. CREATE

Used to create a new table or database object.

Example
CREATE TABLE Students (
    ID INT,
    Name VARCHAR(50),
    Marks INT
);
Purpose

Creates a new table named Students with three columns.

2. ALTER

Used to modify the structure of an existing table.

Example
ALTER TABLE Students
ADD Email VARCHAR(100);
Purpose

Adds a new column Email to the existing table.

You can also:

Modify column datatype
Rename columns
Drop columns
3. DROP

Used to permanently delete a table or database object.

Example
DROP TABLE Students;
Purpose

Deletes:

Table structure
All records inside it

Recovery is difficult unless backup exists.

4. TRUNCATE

Used to remove all rows from a table while keeping the table structure intact.

Example
TRUNCATE TABLE Students;
Purpose

Deletes all data quickly but keeps:

Table name
Columns
Schema

Faster than DELETE because it does not remove rows one by one.
```

```python
Q3. What is DML? Explain INSERT, UPDATE, and DELETE.

DML (Data Manipulation Language) is used to manage data inside tables.

1. INSERT

Adds new data.

INSERT INTO Students
VALUES (1, 'Prince', 95);
2. UPDATE

Modifies existing data.

UPDATE Students
SET Marks = 98
WHERE ID = 1;
3. DELETE

Removes records from a table.

DELETE FROM Students
WHERE ID = 1;
```

```python
Q4. What is DQL? Explain SELECT with an Example.

DQL (Data Query Language) is used to retrieve data from a database.

Main command: SELECT

Example
SELECT * FROM Students;
Purpose

Retrieves all records from the Students table.

Another example:

SELECT Name, Marks
FROM Students
WHERE Marks > 90;
```

```python
Q5. Explain Primary Key and Foreign Key.
Primary Key

A column that uniquely identifies each row in a table.

Properties:

Unique
Cannot contain NULL values
Example
CREATE TABLE Students (
    ID INT PRIMARY KEY,
    Name VARCHAR(50)
);

Here, ID is the Primary Key.

Foreign Key

A column that creates a relationship between two tables.

Example
CREATE TABLE Orders (
    OrderID INT,
    StudentID INT,
    FOREIGN KEY (StudentID) REFERENCES Students(ID)
);

Here:

StudentID is Foreign Key
It references ID from Students table
```

```python
Q6. Python Code to Connect MySQL with Python

Used with MySQL and Python.

Code
import mysql.connector

# Establish connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="your_password",
    database="college"
)

# Create cursor object
cursor = conn.cursor()

# Execute SQL query
cursor.execute("SELECT * FROM Students")

# Fetch and display data
for row in cursor.fetchall():
    print(row)

# Close connection
conn.close()
cursor() Method
Creates a cursor object.
Used to execute SQL commands.
Acts as a bridge between Python and MySQL.
execute() Method
Executes SQL queries.
Can run:
SELECT
INSERT
UPDATE
DELETE

Example:

cursor.execute("SELECT * FROM Students")
```

```python
Q7. Order of Execution of SQL Clauses

Even though we write SQL queries in one order, SQL executes them internally in another order.

Execution Order
FROM
WHERE
GROUP BY
HAVING
SELECT
ORDER BY
LIMIT
Example Query
SELECT Name, COUNT(*)
FROM Students
WHERE Marks > 50
GROUP BY Name
HAVING COUNT(*) > 1
ORDER BY Name;
Internal Execution Flow
FROM → WHERE → GROUP BY → HAVING → SELECT → ORDER BY
```
